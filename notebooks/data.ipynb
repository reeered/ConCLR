```python
with open ("../data/ICDAR2013+2015/labels.txt", "r") as f:
    lines = f.readlines()
```

```python
import os

from PIL import Image
import matplotlib.pyplot as plt

data_dir = '../data/ICDAR2013+2015/train_data'

fig, axs = plt.subplots(2, 4, figsize=(16,7))

k = 10
for i in range(k, k+8):
    img_path, label = lines[i].strip().split('\t')
    img = Image.open(os.path.join(data_dir, img_path))
    t = i - k
    axs[t//4, t%4].imshow(img)
    axs[t//4, t%4].set_title(label)

    
```
