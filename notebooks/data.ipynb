```python
import os

from PIL import Image
import matplotlib.pyplot as plt

# data_dir = '../data/ICDAR2013+2015/train_data'
data_dir = '../data/mnt/ramdisk/max/90kDICT32px'

fig, axs = plt.subplots(2, 4, figsize=(16,7))

k = 10
for i in range(k, k+8):
    # img_path, label = lines[i].strip().split('\t')
    img_path, _ = lines[i].strip().split()
    label = img_path.split('_')[1]
    img = Image.open(os.path.join(data_dir, img_path))
    t = i - k
    axs[t//4, t%4].imshow(img)
    axs[t//4, t%4].set_title(label)

    
```

```python
import sys
sys.path.append("..")
from utils import get_data_loader

test_loader = get_data_loader(data_dir, labels_path, 9, charset_path='../data/charset_36.txt')
```

```python
def plot_images(images, labels):
    fig, axs = plt.subplots(3, 3, figsize=(25,10))
    for t in range(len(images)):
        image = images[t].cpu().clone()
        image = image.squeeze(0).permute(1, 2, 0)
        label = labels[t]
        text = test_loader.dataset.charset.get_text(label, padding=False)
        print(test_loader.dataset.charset.get_text(label))
        # print(test_loader.charset.get_text(label))
        axs[t//3, t%3].imshow(image)
        axs[t//3, t%3].set_title(text)

for i, batch in enumerate(test_loader):
    if i == 1:
        break
    images, labels, view1, view2, labels1, labels2 = batch

    plot_images(images, labels)
    plot_images(view1, labels1)
    plot_images(view2, labels2)

test_loader

```
